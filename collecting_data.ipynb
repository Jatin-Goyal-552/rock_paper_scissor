```python
import cv2
import keyboard 
from time import sleep
```

Collecting train data

```python
vid = cv2.VideoCapture(0)
r=0
s=0
p=0
while True:
    ret,frame=vid.read()
    cv2.rectangle(frame, (50,50), (300,300), (255,0,0),2)
    cv2.imshow("frame",frame)
    frame  = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    img=frame[50:300,50:300]
    if keyboard.is_pressed('r'):
        cv2.imwrite("data/train/rock/"+str(r)+".jpg", img)
        r=r+1
        print("r",r)
        sleep(0.1)
    elif keyboard.is_pressed('p'):
        cv2.imwrite("data/train/paper/"+str(p)+".jpg", img)
        p=p+1
        print("p",p)
        sleep(0.1)
    elif keyboard.is_pressed('s'):
        cv2.imwrite("data/train/scissor/"+str(s)+".jpg", img)
        s=s+1
        print("s",s)
        sleep(0.1)
    if cv2.waitKey(1) & 0xFF == ord('q'): 
        break
  
vid.release() 
cv2.destroyAllWindows()
```
