Importing all the neccessary libraries

```python
import cv2
import tensorflow as tf
from keras.models import load_model
import pickle
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing import image
import numpy as np
import random
import keyboard
from time import sleep
```

 Function to load model and predict the class of image

```python
label_dict = {0:'Paper',1:'Rock',2:'Scissor'}
model = load_model('rps.h5')
def load_image(filename,new_model):
    img = image.load_img(filename,target_size = (100,100),color_mode = "grayscale")
    img = np.array(img)/255
    img = np.expand_dims(img,axis = 0) 
    img = img.reshape(1,100,100,1)
    result = new_model.predict(img)
    result = list(result[0])
    img_index = result.index(max(result))
    return label_dict[img_index]
```

For multiplayer game

```python
vid = cv2.VideoCapture(0)

font = cv2.FONT_HERSHEY_SIMPLEX 
org = (70, 40) 
fontScale = 1.0
color = (255, 255, 0) 
thickness = 3
while True:
    ret,frame=vid.read()
    cv2.rectangle(frame, (20,80), (300,300), (0,0,255),2)
    cv2.rectangle(frame, (350,80), (630,300), (255,0,0),2)
    cv2.putText(frame, "Player 1", (70,30), font,  
                           fontScale, (0,0,255), thickness, cv2.LINE_AA) 
    cv2.putText(frame, "Player 2", (370,30), font,  
                           fontScale, (255,0,0), thickness, cv2.LINE_AA) 
#     frame[80:300,350:630]=frame[80:300,20:300]
#     gray  = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    if keyboard.is_pressed('f'):
        img1=frame[80:300,20:300]
        img2=frame[80:300,350:630]
        cv2.imwrite("a.jpg", img1)
        pred1=load_image('a.jpg',model)
        cv2.imwrite("b.jpg", img2)
        pred2=load_image('b.jpg',model)
        if pred1=="Scissor":
            if pred2=="Paper":
                result="Player 1 wins"
            elif pred2=="Rock":
                result="Player 2 wins"
        elif pred1=="Paper":
            if pred2=="Rock":
                result="Player 1 wins"
            elif pred2=="Scissor":
                result="Player 2 wins"
        elif pred1=="Rock":
            if pred2=="Paper":
                result="Player 2 wins"
            if pred2=="Scissor":
                result="Player 1 wins"
        else:
            result="Tie"
    try:
        cv2.putText(frame, pred1, (75,60), font,  
                           fontScale, (0,0,255), thickness, cv2.LINE_AA) 
        cv2.putText(frame, pred2, (375,60), font,  
                           fontScale, (255,0,0), thickness, cv2.LINE_AA) 
        cv2.putText(frame, result, (150,400), font,  
                           1.5, (0,255,0), 5, cv2.LINE_AA) 
        cv2.imshow("frame",frame)
    except:
        
        cv2.putText(frame, "Press F To Start The Game", (100,70), font,  
                           fontScale, (0,255,0), thickness, cv2.LINE_AA)
        cv2.imshow("frame",frame)
    if cv2.waitKey(1) & 0xFF == ord('q'): 
        break
  
vid.release() 
cv2.destroyAllWindows()
```
